```python
import sqlite3 as sl
import pandas as pd
import datetime
import numpy as np
import pickle

def backtest_2(HYPE, CLOSE, minbuy, minsell,txtname):
    HYPE['diff'] = HYPE.iloc[:, 0].diff(periods=1)
    HYPE['buy'] = (HYPE['diff'] > minbuy)
    HYPE['sell'] = (HYPE['diff'] < minsell)

    CLOSE['upward'] = CLOSE['close'].diff(periods=8)

    CLOSE.index = pd.to_datetime(CLOSE.index)
    HYPE.index = pd.to_datetime(HYPE.index)


    #CLOSE = CLOSE[datetime.date(2021, 9, 1):datetime.date(2021, 11, 1)]
    #HYPE = HYPE[datetime.date(2021, 9, 1):datetime.date(2021, 11, 1)]

    holding = False
    trades = [[],[],[],[]]

    for index, row in HYPE.iterrows():
        if ( (holding == False) and (row['diff'] > minbuy) and (CLOSE.loc[index]['upward'] >= 0)):
            buyprice = CLOSE.loc[index]['close']
            holding = True
            trades[0].append(index)
            trades[1].append(buyprice)

        if ((holding == True) and ((row['diff'] < minsell))):
            sellprice = CLOSE.loc[index]['close']
            trades[2].append(index)
            trades[3].append(sellprice)
            holding = False

    tradeDiff = []
    tradePerc = [1]
    tradeDuration = []

    for x in range(len(trades[0])-1):
        tradeDiff.append(trades[3][x] - trades[1][x])
        tradePerc.append(trades[3][x] / trades[1][x])
        tradeDurationSeconds = (trades[2][x] - trades[0][x]).total_seconds()
        tradeDuration.append( divmod(tradeDurationSeconds, 60)[0] )

    trades_df = pd.DataFrame(trades).transpose()
    trades_df.columns = ["buy_time", "buy_price", "sell_time", "sell_price"]

    tradeReturn = np.prod(tradePerc)
    
    if not tradeDuration:
        averageDuration = 0
    else:
        averageDuration = np.mean(tradeDuration)

        

    str1 = ('id = ' + str(artifact_id) + '    buy = ' + str(minbuy) + '    sell = ' + str(minsell))
    str2 = ('Average duration = ' + str(averageDuration) )
    str3 = ('Trade p/l% = ' + str(tradeReturn) )
    str4 = ('Hold Return = ' + str(CLOSE['close'][-1] / CLOSE['close'][0]) )

    #print(str1)
    #print(str2)
    #print(str3)
    #print(str4)
    #print("\n")
    f = open('{}.txt'.format(txtname), 'a')

    f.write(str1)
    f.write("\n")
    f.write(str2)
    f.write("\n")
    f.write(str3)
    f.write("\n")
    f.write(str4)
    f.write("\n")
    f.write("\n")




    result = pd.Series([artifact_id,
                        minbuy,
                        minsell,
                        len(trades[0]),
                        averageDuration,
                        tradeReturn,
                        CLOSE['close'][-1] / CLOSE['close'][0]],
                        index=['id',
                        'minbuy', 
                        'minsell', 
                        'trades',
                        'av_time', 
                        'bot_return', 
                        'hold_return'])

    # result = {
    # "id": artifact_id,
    # "minbuy": minbuy,
    # "minsell": minsell,
    # "av_time": averageDuration,
    # "bot_return": tradeReturn,
    # "hold_return": CLOSE['close'][-1] / CLOSE['close'][0],
    # "trades": pd.DataFrame(trades)
    # }
    return result, trades_df

def backtest_flat(HYPE, CLOSE, minbuy, minsell,txtname):
    HYPE['diff'] = HYPE.iloc[:, 0].diff(periods=1)
    CLOSE['upward'] = CLOSE['close'].diff(periods=8)

    CLOSE.index = pd.to_datetime(CLOSE.index)
    HYPE.index = pd.to_datetime(HYPE.index)


    #CLOSE = CLOSE[datetime.date(2021, 9, 1):datetime.date(2021, 11, 1)]
    #HYPE = HYPE[datetime.date(2021, 9, 1):datetime.date(2021, 11, 1)]

    holding = False
    trades = [[],[],[],[]]

    for index, row in HYPE.iterrows():
        if ( (holding == False) and (row['hype'] > minbuy) and (CLOSE.loc[index]['upward'] >= 0)):
            buyprice = CLOSE.loc[index]['close']
            holding = True
            trades[0].append(index)
            trades[1].append(buyprice)

        if ((holding == True) and (row['hype'] < minsell)):
            sellprice = CLOSE.loc[index]['close']
            trades[2].append(index)
            trades[3].append(sellprice)
            holding = False

    tradeDiff = []
    tradePerc = [1]
    tradeDuration = []

    for x in range(len(trades[0])-1):
        tradeDiff.append(trades[3][x] - trades[1][x])
        tradePerc.append(trades[3][x] / trades[1][x])
        tradeDurationSeconds = (trades[2][x] - trades[0][x]).total_seconds()
        tradeDuration.append( divmod(tradeDurationSeconds, 60)[0] )

    trades_df = pd.DataFrame(trades).transpose()
    trades_df.columns = ["buy_time", "buy_price", "sell_time", "sell_price"]

    tradeReturn = np.prod(tradePerc)
    
    if not tradeDuration:
        averageDuration = 0
    else:
        averageDuration = np.mean(tradeDuration)

        

    str1 = ('id = ' + str(artifact_id) + '    buy = ' + str(minbuy) + '    sell = ' + str(minsell))
    str2 = ('Average duration = ' + str(averageDuration) )
    str3 = ('Trade p/l% = ' + str(tradeReturn) )
    str4 = ('Hold Return = ' + str(CLOSE['close'][-1] / CLOSE['close'][0]) )

    #print(str1)
    #print(str2)
    #print(str3)
    #print(str4)
    #print("\n")
    f = open('{}.txt'.format(txtname), 'a')

    f.write(str1)
    f.write("\n")
    f.write(str2)
    f.write("\n")
    f.write(str3)
    f.write("\n")
    f.write(str4)
    f.write("\n")
    f.write("\n")




    result = pd.Series([artifact_id,
                        minbuy,
                        minsell,
                        len(trades[0]),
                        averageDuration,
                        tradeReturn,
                        CLOSE['close'][-1] / CLOSE['close'][0]],
                        index=['id',
                        'minbuy', 
                        'minsell', 
                        'trades',
                        'av_time', 
                        'bot_return', 
                        'hold_return'])

    # result = {
    # "id": artifact_id,
    # "minbuy": minbuy,
    # "minsell": minsell,
    # "av_time": averageDuration,
    # "bot_return": tradeReturn,
    # "hold_return": CLOSE['close'][-1] / CLOSE['close'][0],
    # "trades": pd.DataFrame(trades)
    # }
    return result, trades_df
```

```python
con = sl.connect("crypto_database.db")
cur = con.cursor()
```

```python
open('results.txt', 'w').close()
artifacts = pd.read_sql_query('SELECT id, name FROM artifact', con)
results = pd.DataFrame(columns=['id','minbuy', 'minsell', 'trades', 'av_time', 'bot_return', 'hold_return'])
trade_data = pd.DataFrame(columns=['id', 'minbuy', 'minsell', "buy_time", "buy_price", "sell_time", "sell_price"])

for index, row in artifacts.iterrows():
    artifact_id = row[0]
    name = row[1]
    
    closeString = 'SELECT date, close FROM artifact_price WHERE artifact_id="{}"'.format(artifact_id)
    hypeString = 'SELECT date, hype FROM artifact_trend WHERE artifact_id="{}"'.format(artifact_id)
    
    value = pd.read_sql_query(closeString, con)
    trend = pd.read_sql_query(hypeString, con)

    if value.empty or trend.empty:
        print(str(index) + "failed due to empty")
        continue

    value = value.set_index('date')
    trend = trend.set_index('date')

    half_way = '2021-08-28 21:00:00'
    try:
        # value = value[trend.index[0]:half_way]
        # trend = trend[value.index[0]:half_way]
        value = value[trend.index[0]:]
        trend = trend[value.index[0]:]
        value = value[half_way:]
        trend = trend[half_way:]
    except:
        print(str(index) + "failed due to 28-08")
        continue

    for minbuy in [0, 5, 10, 20, 30]:
        for minsell in [-0, -5, -10, -20, -30]:
            result, trades = backtest_2(trend, value, minbuy, minsell)
            results = results.append(result,ignore_index=True)
            trades.insert(0, 'id', artifact_id)
            trades.insert(1, 'minbuy', minbuy)
            trades.insert(2, 'minsell', minsell)
            trade_data = trade_data.append(trades, ignore_index=True)

results.to_pickle("./results_flat_2.pkl")
trade_data.to_pickle("./trade_flat_2.pkl")
#pickle.dump( trade_data, open( "./trade_data_04_12_second.pkl", "wb" ) )
```
