```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates

plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (40,20)

pytrends = TrendReq(hl='en-GB')

all_keywords = [['hertz network']]
keywords = []

timeframes = ['today 5-y', 'today 12-m', 'today 3-m', 'today 1-m',
              'now 7-d', 'now 1-d', 'now 4-H']
cat = '0'
geo = ''
gprop = ''


def graph_trends():
    for kw in range(len(all_keywords)):
        pytrends.build_payload(all_keywords[kw],
                               cat,
                               timeframes[5],
                               geo,
                               gprop)
        data = pytrends.interest_over_time()
        plt.figure()
        ax1 = plt.subplot2grid((3,2), (0,0), rowspan = 1, colspan = 1)
        ax2 = plt.subplot2grid((3,2), (1,0), rowspan = 1, colspan = 1)
        ax1.plot(data, label = kw)
        ax2.plot(data.rolling(10).mean(), label = kw)

        
        

    
```

```python
graph_trends()

```
